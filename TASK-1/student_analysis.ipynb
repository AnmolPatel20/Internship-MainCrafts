#### Importing the required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Visulation
sns.set(style="whitegrid")
```

#### Loading The Dataset and also seperate the attributes 

```python
mat_df = pd.read_csv("student-mat.csv", sep=';')
por_df = pd.read_csv("student-por.csv", sep=';')
```

```python
#Display the Shape of the dataset (rows and columns)

print("Math dataset shape:", mat_df.shape)
print("Portuguese dataset shape:", por_df.shape)
```

```python
#Preview the DATA
mat_df.head()
```

```python
#Displaying all the atrributes using - .columns
mat_df.columns
```

#### Among These we gonna use "G3" final grade.. which is our main target 

```python
# Dataset structure and datatypes
mat_df.info()
```

```python
# Statistical summary of attributs
mat_df.describe()
```

#### DATA CLEANING

```python
# Check missing values
mat_df.isnull().sum()
```

```python
# Check duplicate rows
mat_df.duplicated().sum()
```

```python
# Remove duplicates... doing on a safer side to check even if none exists
mat_df = mat_df.drop_duplicates()
```

##### So We can say that our Dataset is Cleaned and ready for the use 

#### DATA ANALYSIS

```python
# Average final grade (QUE-1)
mat_df["G3"].mean()
```

```python
# Students scoring above 15 (QUE-2)
mat_df[mat_df["G3"] > 15].shape[0]
```

```python
# Correlation between study time and final grade (QUE-3)
mat_df["studytime"].corr(mat_df["G3"])
```

```python
# Gender-wise average performance either MALE or FEMALE (QUE-4)
mat_df.groupby("sex")["G3"].mean()
```

#### DATA VISUALIZATION

```python
# Histogram of final grades  (QUE-1)
plt.figure(figsize=(8,5))
plt.hist(mat_df["G3"], bins=10, edgecolor="black")
plt.title("Distribution of Final Grades (Math)")
plt.xlabel("Grade")
plt.ylabel("Frequency")
plt.show()
```

```python
# Study time vs final grade   (QUE-2)
plt.figure(figsize=(8,5))
plt.scatter(mat_df["studytime"], mat_df["G3"])
plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade")
plt.show()
```

```python
# Gender-wise comparison   (QUE-3)
mat_df.groupby("sex")["G3"].mean().plot(kind="bar")
plt.title("Average Final Grade by Gender")
plt.ylabel("Average G3")
plt.show()
```

Additional note: there are several (382) students that belong to both datasets 
These students can be identified by searching for identical attributes
that characterize each student, as shown in the annexed R file.
##### ~This things given in student.txt file 

##### Columns used to uniquely identify students (as per R file) Here is the Screenshot
![My Image](R_Content_Screenshot.png)

```python
id_cols = [
    "school","sex","age","address","famsize","Pstatus",
    "Medu","Fedu","Mjob","Fjob","reason",
    "nursery","internet"
]
```

```python
# Merge Math and Portuguese datasets
merged_df = pd.merge(
    mat_df,
    por_df,
    on=id_cols,
    suffixes=("_math", "_por")
)

merged_df.shape
```

#### The merged dataset contains **382 common students** matching the result that is obtained from the reference R file.

```python
# Compare average final grades
merged_df[["G3_math", "G3_por"]].mean()
```

```python
# Calculate average final grades and comapre by visualizing
merged_df[["G3_math", "G3_por"]].mean().plot(kind="bar")

plt.title("Average Final Grades: Math vs Portuguese")
plt.ylabel("Average Grade")
plt.show()
```
