##### Importing the required Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

##### Loading the Dataset 

```python
df = pd.read_csv("TitanicDataset.csv")
```

```python
df.head()  #check that dataset is loaded properly or not
```

### Dataset Overview

```python
df.info()
```

```python
df.isnull().sum()
```

### Initial observation

Dataset contains 1309 Rows and 28 Columns <br>
Too many colums named as 'zero', 'zero.1' are present; These columns are not required and not meaningful for EDA <br>
Column '2urvived' is written; These represents the Survival Status<br>
Column 'Embarked' has 2 missing values <br>
Column 'Sex' is already encoded as numeric<br>

1. Rename Target Column : '2urvived' -> 'Survived'

```python
df.rename(columns={'2urvived': 'Survived'}, inplace=True)
df.columns
```

2. Drop Down Irrelevent Columns [zero, zero.1, zero.2, etc] also Passengerid be dropped as it is only an identifier and this values do not add value to survival analyis 

```python
zero_cols = [col for col in df.columns if col.startswith('zero')]  #drop all the columns stating with zero
df.drop(columns=zero_cols + ['Passengerid'], inplace=True)  #both Passengerid and zeros are dropped
df.columns
```

3. Handling Missing Values in the Column [Embarked], All those missing values are filled with most occuring number from the column so for that we are using (mode)

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.isnull().sum()
```

After removing all the irrelevent columns and handled missing values. Our Dataset is now ready For EDA (Exploratory Data Analysis)

```python
df.head()
```

#### Now we see that how many numbers of passengers survived v/s those who did not survive

```python
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Number of Passengers")
plt.show()
```

## Survival Analysis by Gender 
##### Comparing the survival rates of Male and Female

```python
df.groupby('Sex')['Survived'].mean()
```

```python
sns.barplot(x='Sex', y='Survived', data=df)
plt.title("Survival Rate by Gender")
plt.xlabel("Sex (0 = Female, 1 = Male)")
plt.ylabel("Survival Rate")
plt.show()
```

#### Answer-1: As per Analysis, I found that Male Survived More than Female and ploted the bar chart of survival by gender

## Survival Analysis by Passenger Class
##### Survival Chances of Passenger Class during Titanic Collide

```python
df.groupby('Pclass')['Survived'].mean()
```

```python
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.show()
```

#### Answer-2: As per Analysis, Passenger class  significantly affected survival chances
#### First Class has highest survival rate, followed by second class, where third class passenger had the lowest survival rate

## Survival Analysis by Age Group
##### Calculate the Survival rate for each age group and understanding how age get affected survival chances
##### Creating Age Group, Divide Passengers into Different age Categories.

```python
bins = [0, 12, 18, 40, 60, 100]
labels = ['Child', 'Teen', 'Adult', 'Middle Aged', 'Senior']

df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)
df['AgeGroup'].value_counts()
```

```python
df.groupby('AgeGroup', observed=True)['Survived'].mean()
```

```python
sns.barplot(x='AgeGroup', y='Survived', data=df)
plt.title("Survival Rate by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Survival Rate")
plt.show()
```

#### Answer-3: As per the Analysis, Survived chances varied acreoss age groups
#### Younger Passengers having Better Survival Outcome , Senior had the Lowest Survival Rate

### Age Distributions of Passengers
Displaying Histogram that shiows the distribution of Passengers ages on the Titanic

```python
plt.hist(df['Age'], bins=20)
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.show()
```

##### The age distribution shows that most passengers were in the 20â€“30 age range, indicating this was the most common age group.
